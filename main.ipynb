```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
numerical_data.remove('index')
numerical_data.remove('Unnamed: 0')
```

```python
numerical_data
```

Histogram

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(14,10))

for i, colum in enumerate(numerical_data):
    row = i//3
    col = i%3
    ax = axes[row,col]
    sns.histplot(data = train_df[colum],ax = ax)

plt.show()
```

Density Plot

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(14,10))

for i, colum in enumerate(numerical_data):
    row = i//3
    col = i%3
    ax = axes[row,col]
    sns.kdeplot(data = train_df[colum],ax = ax,fill=True)

plt.show()
```

Box and Whisker Plots

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(14,10))

for i, colum in enumerate(numerical_data):
    row = i//3
    col = i%3
    ax = axes[row,col]
    sns.boxplot(data = train_df[colum],ax = ax)
    ax.set_xlabel(colum)
fig.tight_layout()

plt.show()
```
